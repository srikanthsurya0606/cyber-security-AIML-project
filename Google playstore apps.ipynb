```python
df=pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['clarity'],df['price'])
plt.title("clarity vs price")
plt.xlabel("clarity")
plt.ylabel("price")
plt.show()
```

```python
plt.bar(df['clarity'],df['price'])
plt.title("clarity vs price")
plt.xlabel("clarity")
plt.ylabel("price")
plt.show()
```

```python
plt.plot(df['price'],df['clarity'])
plt.title("price vs clarity")
plt.xlabel("price")
plt.ylabel("clarity")
plt.show()
```

```python
plt.hist(df['price'],bins=4)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.suptitle("final report")
```

```python
plt.subplot(2,2,1)
plt.scatter(df['clarity'],df['price'])
plt.title("clarity vs price")
plt.xlabel("clarity")
plt.ylabel("price")
plt.subplot(2,2,2)
plt.bar(df['clarity'],df['price'])
plt.title("clarity vs price")
plt.xlabel("clarity")
plt.ylabel("price")
plt.subplot(2,2,3)
plt.plot(df['price'],df['clarity'])
plt.title("price vs clarity")
plt.xlabel("price")
plt.ylabel("clarity")
plt.subplot(2,2,4)
plt.hist(df['price'],bins=4)
plt.tight_layout()
plt.savefig("srikanth.png")
plt.show()

```
